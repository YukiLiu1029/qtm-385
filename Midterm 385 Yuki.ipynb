````python
class Matrix:
    def __init__(self, num, nrow):
        if(len(num)%nrow == 0):
            matrix = []
            self.num = num
            self.nrow = nrow
            ncol= int(len(num)/(nrow))
            k = 0
            for i in range(nrow):
                matrix.append(num[k:k+ncol])
                k = k+ncol
            self.mat = matrix
            self.ncol = ncol
            self.dim = [nrow,ncol]
        pass
    
    # Print represetation
    def __str__(self):
        s = ""
        m = 0
        for i in range(self.nrow):
            s+='['
            for j in range(self.ncol-1):
                s+=str(self.num[m])+','
                m+=1
            s+=str(self.num[m])+']\n'
            m+=1
        s = s[:-1]
        return s
    
    # Representation
    def __repr__(self):
        return "Matrix dimension [{row},{col}]".format(row=self.nrow,col=self.ncol)
    
    # Test to see if square matrix
    def is_square(self):
        return (self.nrow == self.ncol)
        pass
        
    # Addition 
    def __add__(self, other):
        if(self.dim == other.dim):
            new_num = []
            for i in range(len(self.num)):
                new_num.append(self.num[i]+other.num[i])
            self.num = new_num
            return self
        else:
            raise ValueError('Error. The matrices do not conform.')
        pass
    
    # Subtraction
    def __sub__(self, other):
        if(self.dim == other.dim):
            new_num = []
            for i in range(len(self.num)):
                new_num.append(self.num[i]-other.num[i])
            self.num = new_num
            return self
        else:
            raise ValueError('Error. The matrices do not conform.')
        pass
        
    # Scalar multiplication 
    def __mul__(self, num):
        new_num = []
        for i in range(len(self.num)):
            new_num.append((self.num[i])*num)
        self.num = new_num
        return self
        pass
    
    # Extract row
    def exrow(self, i):
        if(i<=self.nrow-1):
            return self.mat[i]
        else:
            return 'Error. Invalid row.'
        pass
    
    # Extract column
    def excol(self, i):
        if(i<=self.ncol-1):
            col = []
            for j in range(self.nrow):
                col.append(self.mat[j][i])
            return col
        else:
            return 'Error. Invalid column.'
        pass
    
    # Matrix multiplication 
    def __pow__(self, other):
        if(self.ncol == other.nrow):
            new_num=[]
            for j in range(self.nrow):
                for i in range(other.ncol):
                    s=0
                    for k in range(other.nrow):
                        s+=self.mat[j][k]*other.mat[k][i]
                    new_num.append(s)
            new_mat = Matrix(new_num,self.nrow)
            return new_mat
        else:
            return 'Error. Matrices do not conform for multiplication.'
        pass
    
    # Extract submatrix for minor 
    def subm(self, row, col):
        new_num = []
        for j in range(self.nrow):
            if(j!=row):
                for i in range(self.ncol):
                    if(i!= col):
                        new_num.append(self.mat[j][i])
        new_matrix = Matrix(new_num,(self.ncol)-1)
        return new_matrix
        pass
    
    # Determinant 
    def det(self):
        if(self.is_square()):
            determinant = 0
            if(self.nrow == 2):
                return (self.num[0]*self.num[3]-self.num[2]*self.num[1])
            else:
                for i in range(self.ncol):
                    determinant+=((-1)**(i))*(self.mat[0][i])*(self.subm(0,i).det())
                return determinant
        else:
            return 'We cannot calculate the determinant since the matrix is not square.'
        pass
    
    # Transpose
    def t(self):
        new_num=[]
        new_nrow=self.ncol
        for i in range(self.ncol):
            for j in range(self.nrow):
                new_num.append(self.excol(i)[j])
        new_mat = Matrix(new_num, new_nrow)
        return new_mat
        pass
    
    # Cofactor matrix 
    def cofm(self):
        if(self.is_square()):
            new_num =[]
            for j in range(self.nrow):
                for k in range(self.ncol):
                    new_num.append((-1)**(j+k)*(self.subm(j,k)).det())
            cof_mat = Matrix(new_num,self.ncol)
            return cof_mat
        pass
    
    # Inverse matrix 
    def inv(self):
        if(self.det()!=0):
            new_num = (self.cofm().t()*(1/self.det())).num
            inv_mat = Matrix(new_num,self.ncol)
            return inv_mat
        else:
            return 'Error, since matrix is singular or not square.'
        pass
    
    def regcof(self,other):
        if((self.nrow == other.nrow) & (self.ncol == self.ncol)):
            return ((((self.t())**self).inv())**self.t())**other
        pass
           
````
